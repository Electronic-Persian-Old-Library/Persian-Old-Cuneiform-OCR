```python
char_map = {
"êé†":"a",
"êé°":"ii", 
"êé¢":"≈´", 
"êé£":"ka",  
"êé§":"ku",  
"êé•":"ga",  
"êé¶":"gu",  
"êéß":"x", 
"êé®":"c",  
"êé©":"ja",  
"êé™":"ji",  
"êé´":"ta",  
"êé¨":"tu",  
"êé≠":"da",  
"êéÆ":"di",  
"êéØ":"du",
"êé∞":"th",  
"êé±":"p", 
"êé≤":"b", 
"êé≥":"f", 
"êé¥":"na",
"êéµ":"nu", 
"êé∂":"ma",
"êé∑":"mi", 
"êé∏":"mu",
"êéπ":"y", 
"êé∫":"va",  
"êéª":"vi",  
"êéº":"ra", 
"êéΩ":"ru", 
"êéæ":"r",
"êéø":"s",  
"êèÄ":"z",  
"êèÅ":"sh", 
"êèÇ":"ss",
"êèÉ":"h",  
"êèê": " ",
"êèë":"1",
"êèí":"2",
"êèì":"10", 
"êèî":"20",
"êèï":"100",
"êèã":"king",
"êèè":"earth",
"êèåêèç":"country",
"êèé":"God",
"êèàêèâ":"Auramazda",
"êèä":"Auramazdaha",

}
```

```python
def replace_characters_with_custom(original_string, char_map):
    # Create a new string by replacing each character according to the dictionary
    new_string = ''.join(char_map.get(char, char)
    for char in original_string)
    return new_string
```

```python
original_string = input('Enter your Cuneiform: ')   #example:êé≠êé†êéæêéπêé∫êé¢êèÅêèêêèã
new_string = replace_characters_with_custom(original_string, char_map)
print('Transliteration:',new_string) 
```
